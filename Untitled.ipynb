```python
import pandas as pd
from itertools import combinations

# Load the orders data into a pandas dataframe
orders_df = pd.read_csv( "D:\\Academics\\Datahack\\jetson-sample-data.csv")

# Group the orders by order ID and aggregate the items into a list
order_items_df = orders_df.groupby('order_id')['item_name'].agg(list).reset_index()

# Create a list of all pairs of items that were ordered together
item_pairs = []
for item_list in order_items_df['item_name']:
    pairs = combinations(item_list, 2)
    item_pairs.extend(pairs)

# Count the number of times each item pair appears in the list
item_pair_counts = pd.Series(item_pairs).value_counts()

# Print the top 10 item pairs ordered together
print(item_pair_counts.head(10))
```

```python

import pandas as pd
from itertools import combinations
from collections import Counter

# Create sample data
orders = pd.read_csv( "D:\\Academics\\Datahack\\jetson-sample-data.csv")

# Get all pairs of items
item_pairs = list(combinations(set(orders['item_name']), 2))

# Get frequency counts for each item pair
freq_counts = Counter()

for order_id, group in orders.groupby('order_id'):
    items = set(group['item_name'])
    for pair in combinations(items, 2):
        freq_counts[pair] += 1

# Create frequency dataframe
freq_df = pd.DataFrame.from_dict(freq_counts, orient='index', columns=['frequency'])
freq_df.index = pd.MultiIndex.from_tuples(freq_df.index, names=['item1', 'item2'])
freq_df = freq_df.reset_index()

# Sort by frequency in descending order
freq_df = freq_df.sort_values('frequency', ascending=False)
freq_df.head(100)
# Print recommendations for each item
# for item in set(orders['item_name']):
#     recommendations = freq_df[(freq_df['item1'] == item) | (freq_df['item2'] == item)]
#     recommendations = recommendations[['item1', 'item2', 'frequency']]
#     print(f"Recommendations for {item}:")
#     print(recommendations)
```
